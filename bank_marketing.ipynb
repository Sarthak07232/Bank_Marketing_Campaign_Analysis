#   4. Campaign Effectiveness:

###  Does the duration of the call increase the chances of a successful subscription?

```python
f=df.groupby('duration')['y'].value_counts(normalize=True)
```

```python
f.plot(kind='box',figsize=(10,5))
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.title('call duration & chances of success', fontsize=14, fontweight='bold')
plt.xlabel('Subsciption', fontsize=12)
plt.ylabel('duration', fontsize=12)
plt.show()
```

```python
sb.scatterplot(x=df.duration,y=df.y)
plt.show()
```

###  Which contact methods (e.g., telephone, cellular) are most effective?


```python
e=df.groupby('contact')['y'].value_counts(normalize=True)
```

```python
e.plot(kind='bar',figsize=(10,5),color=['black','red'])  #celllar is most effective 4.2
```

###  How many contacts are usually needed before a client subscribes?


```python
p=df.groupby('y')['campaign'].mean()
```

```python
p.plot(kind='bar',figsize=(10,5),color=['black','red'])
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.title('contacts and subscription ratio', fontsize=14, fontweight='bold')
plt.xlabel('y', fontsize=12)
plt.ylabel('campaign', fontsize=12)
plt.show()
```
